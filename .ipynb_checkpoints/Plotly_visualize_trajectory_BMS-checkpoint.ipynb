```python
import sys
import numpy as np 
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from mcmc import *
from parallel import *
from fit_prior import read_prior_par

import plotly.graph_objects as go
```

## Import data and building dataframe

```python
filename = './Results/2107_Investigate_complexity_factor/Nitrogen_20_10_4/All_Models175871.txt'
models_df = pd.read_csv(filename, delimiter=',')
models_df
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=np.arange(len(models_df.index)),
                         y=np.array(models_df['Description length']).astype(float), 
                         opacity=0.5,
                         marker=dict(size=10,
                                     line=dict(color='White',
                                               width=0.5)),
                         hovertemplate='Model: %{text}' 
                         '<br>MCMC step: %{x}' 
                         '<br>Description length: %{y:4.4f}'
                         '<br>a<sub>0</sub>: %{customdata[0]:4.4f}'
                         '<br>a<sub>1</sub>: %{customdata[1]:4.4f}'
                         '<br>a<sub>2</sub>: %{customdata[0]:4.4f}'
                         '<br>a<sub>3</sub>: %{customdata[0]:4.4f}',                         
                         text=models_df['Model'],
                         customdata=models_df[['_a0_','_a1_','_a2_','_a3_']],
                         name='Trajectory',
                         showlegend=True
                        )
             )

fig.update_layout(title='Trajectory of models by BMS on Nitrogen dataset - Setting: axiom1_weight = 20, axiom2_weight = 10, complexity_factor = 4',
                  xaxis_title='MCMC step',
                  yaxis_title='Description length',
                  autosize=False,
                  width=1600,
                  height=800,
                 )

fig.write_html( './Results/2107_Investigate_complexity_factor/Nitrogen_20_10_4/MDL_N2_20104_100000.html')
```
